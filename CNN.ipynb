```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt

# 데이터 로드 및 전처리
print("=== MNIST 데이터 로드 ===")
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
print(f"Train data shape: {x_train.shape}")
print(f"Test data shape: {x_test.shape}")

# 데이터 정규화 및 차원 확장
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
x_train = np.expand_dims(x_train, axis=-1)  # (28, 28, 1)
x_test = np.expand_dims(x_test, axis=-1)

# 원핫 인코딩
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)

print(f"Processed train data shape: {x_train.shape}")
print(f"Processed test data shape: {x_test.shape}")

# 1. 기본 CNN 모델 구현 (강의자료 기준)
def create_basic_cnn():
    model = keras.Sequential([
        # Conv1: 3x3 kernel, 32 filters
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1), padding='same'),
        layers.MaxPooling2D((2, 2)),

        # Conv2: 3x3 kernel, 64 filters
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D((2, 2)),

        # Flattening
        layers.Flatten(),

        # FC1: Fully Connected Layer
        layers.Dense(64, activation='relu'),

        # Output Layer
        layers.Dense(10, activation='softmax')
    ])
    return model

# 2. 개선된 CNN 모델 (Conv3 + FC2 추가)
def create_improved_cnn():
    model = keras.Sequential([
        # Conv1: 3x3 kernel, 32 filters
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1), padding='same'),
        layers.MaxPooling2D((2, 2)),

        # Conv2: 3x3 kernel, 64 filters
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D((2, 2)),

        # Conv3: 3x3 kernel, 128 filters (추가)
        layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D((2, 2)),

        # Flattening
        layers.Flatten(),

        # FC1: Fully Connected Layer
        layers.Dense(128, activation='relu'),

        # FC2: 추가 Fully Connected Layer
        layers.Dense(64, activation='relu'),

        # Output Layer
        layers.Dense(10, activation='softmax')
    ])
    return model

# 모델 생성
print("\n=== 모델 생성 ===")
basic_model = create_basic_cnn()
improved_model = create_improved_cnn()

# 모델 컴파일
basic_model.compile(optimizer='adam',
                   loss='categorical_crossentropy',
                   metrics=['accuracy'])

improved_model.compile(optimizer='adam',
                      loss='categorical_crossentropy',
                      metrics=['accuracy'])

# 모델 구조 출력
print("\n=== 기본 CNN 모델 구조 ===")
basic_model.summary()

print("\n=== 개선된 CNN 모델 구조 ===")
improved_model.summary()

# 모델 학습
print("\n=== 기본 CNN 모델 학습 ===")
basic_history = basic_model.fit(x_train, y_train,
                               batch_size=32,
                               epochs=10,
                               validation_data=(x_test, y_test),
                               verbose=1)

print("\n=== 개선된 CNN 모델 학습 ===")
improved_history = improved_model.fit(x_train, y_train,
                                     batch_size=32,
                                     epochs=10,
                                     validation_data=(x_test, y_test),
                                     verbose=1)

# 모델 평가
print("\n=== 모델 평가 결과 ===")
basic_test_loss, basic_test_acc = basic_model.evaluate(x_test, y_test, verbose=0)
improved_test_loss, improved_test_acc = improved_model.evaluate(x_test, y_test, verbose=0)

print(f"기본 CNN 모델:")
print(f"  - Test Loss: {basic_test_loss:.4f}")
print(f"  - Test Accuracy: {basic_test_acc:.4f}")

print(f"\n개선된 CNN 모델 (Conv3 + FC2 추가):")
print(f"  - Test Loss: {improved_test_loss:.4f}")
print(f"  - Test Accuracy: {improved_test_acc:.4f}")

print(f"\n정확도 개선: {improved_test_acc - basic_test_acc:.4f}")

# 학습 과정 시각화
plt.figure(figsize=(15, 5))

# 정확도 비교
plt.subplot(1, 3, 1)
plt.plot(basic_history.history['accuracy'], label='Basic Model (Train)')
plt.plot(basic_history.history['val_accuracy'], label='Basic Model (Test)')
plt.plot(improved_history.history['accuracy'], label='Improved Model (Train)')
plt.plot(improved_history.history['val_accuracy'], label='Improved Model (Test)')
plt.title('Model Accuracy Comparison')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

# 손실 비교
plt.subplot(1, 3, 2)
plt.plot(basic_history.history['loss'], label='Basic Model (Train)')
plt.plot(basic_history.history['val_loss'], label='Basic Model (Test)')
plt.plot(improved_history.history['loss'], label='Improved Model (Train)')
plt.plot(improved_history.history['val_loss'], label='Improved Model (Test)')
plt.title('Model Loss Comparison')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# 최종 정확도 막대 그래프
plt.subplot(1, 3, 3)
models = ['Basic CNN', 'Improved CNN\n(+Conv3, +FC2)']
accuracies = [basic_test_acc, improved_test_acc]
colors = ['skyblue', 'lightcoral']

bars = plt.bar(models, accuracies, color=colors)
plt.title('Final Test Accuracy Comparison')
plt.ylabel('Accuracy')
plt.ylim(0.95, 1.0)

# 막대 위에 정확도 값 표시
for bar, acc in zip(bars, accuracies):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.001,
             f'{acc:.4f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()

# 예측 결과 시각화
print("\n=== 예측 결과 비교 ===")
# 테스트 데이터에서 몇 개 샘플 선택
sample_indices = np.random.choice(len(x_test), 6, replace=False)
sample_images = x_test[sample_indices]
sample_labels = np.argmax(y_test[sample_indices], axis=1)

# 예측 수행
basic_predictions = basic_model.predict(sample_images)
improved_predictions = improved_model.predict(sample_images)

basic_pred_labels = np.argmax(basic_predictions, axis=1)
improved_pred_labels = np.argmax(improved_predictions, axis=1)

# 결과 시각화
plt.figure(figsize=(15, 8))
for i in range(6):
    plt.subplot(2, 6, i+1)
    plt.imshow(sample_images[i].reshape(28, 28), cmap='gray')
    plt.title(f'True: {sample_labels[i]}')
    plt.axis('off')

    plt.subplot(2, 6, i+7)
    plt.text(0.1, 0.8, f'Basic CNN: {basic_pred_labels[i]}', fontsize=12,
             color='blue' if basic_pred_labels[i] == sample_labels[i] else 'red')
    plt.text(0.1, 0.5, f'Improved CNN: {improved_pred_labels[i]}', fontsize=12,
             color='blue' if improved_pred_labels[i] == sample_labels[i] else 'red')
    plt.text(0.1, 0.2, f'Actual: {sample_labels[i]}', fontsize=12, color='black')
    plt.xlim(0, 1)
    plt.ylim(0, 1)
    plt.axis('off')

plt.suptitle('Prediction Results Comparison', fontsize=16)
plt.tight_layout()
plt.show()

# 모델 파라미터 수 비교
print("\n=== 모델 파라미터 수 비교 ===")
basic_params = basic_model.count_params()
improved_params = improved_model.count_params()

print(f"기본 CNN 모델 파라미터 수: {basic_params:,}")
print(f"개선된 CNN 모델 파라미터 수: {improved_params:,}")
print(f"파라미터 증가량: {improved_params - basic_params:,}")

print("\n=== 과제 완료 ===")
print("1. 기본 CNN 모델 구현 및 평가 완료")
print("2. Conv3, FC2 레이어 추가한 개선 모델 구현 완료")
print("3. 두 모델의 성능 비교 완료")

# 결론
if improved_test_acc > basic_test_acc:
    print(f"\n결론: 개선된 모델이 기본 모델보다 {improved_test_acc - basic_test_acc:.4f} 높은 정확도를 보입니다.")
else:
    print(f"\n결론: 개선된 모델의 성능이 기본 모델보다 {basic_test_acc - improved_test_acc:.4f} 낮습니다.")
    print("   (오버피팅이나 모델 복잡도 증가로 인한 것일 수 있습니다)")
```
